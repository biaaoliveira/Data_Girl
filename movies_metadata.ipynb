```python
import pandas as pd
```

```python
df = pd.read_csv('movies_metadata.csv', sep = ',', low_memory= False)
```

```python
print(df)
```

```python
colunas_utilizadas = ['title', 'budget', 'popularity', 'release_date', 'revenue']
df = df[colunas_utilizadas]
```

```python
print(df)
```

```python
df = df.rename(columns = {'title': 'Titulo',
                          'budget': 'Orçamento',
                          'popularity': 'Popularidade',
                          'release_date': 'Data_lançamento',
                          'revenue': 'Receita'
                          })
```

```python
print(df)
```

```python
df['Orçamento'] = pd.to_numeric(df['Orçamento'], errors = 'coerce')
df['Popularidade'] = pd.to_numeric(df['Popularidade'], errors = 'coerce')
df['Receita'] = pd.to_numeric(df['Receita'], errors = 'coerce')
```

```python
df = df.dropna()
```

```python
df
```

```python
df['Data_lançamento'] = pd.to_datetime(df['Data_lançamento'], errors = 'coerce')
```

```python
df['Ano_Lançamento'] = df ['Data_lançamento'].dt.year
```

```python
print(df)
```

```python
df.to_csv('filmes_tratados.csv', sep =';')
```
